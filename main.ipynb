```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('HR-Employee-Attrition.csv')

# View the first few rows of the dataset
print(df.head())
```

```python
# Check data info and null values
print(df.info())
print(df.isnull().sum())

# Check unique values for each column (helps to understand categorical data)
print(df.nunique())
```

```python
# Drop irrelevant columns
df = df.drop(['EmployeeNumber', 'EmployeeCount', 'Over18', 'StandardHours'], axis=1)
```

```python
print(df.info())
```

```python
# Plot the distribution of Attrition
sns.countplot(x='Attrition', data=df)
plt.title('Attrition Distribution')
plt.show()
```

```python
# Plot Attrition against OverTime
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title('Attrition vs OverTime')
plt.show()
```

```python
# Plot JobSatisfaction against Attrition
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)
plt.title('Attrition vs Job Satisfaction')
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
